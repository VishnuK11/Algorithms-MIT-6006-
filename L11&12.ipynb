# Newton Method - Square root

```python
from math import pow,log
import sys
```

```python
a=2*pow(10,64) # The number
# Theory; X(i+1)= (Xi +(a/xi))/2
x=[1]  # X0
d=64 # Number of digits
counter=0
i=0
while i<=log(d,2):
    temp=(x[i]+(a/x[i]))/2
    x.append(temp)
    print("{:10.256f}".format(temp),"digits",pow(2,i))
    i+=1
```

```python
# This is a heap without balancing and without binary sort..
# Condition of heap: value(parent) < value(child 1) and value(parent) < value(child 2)

class binary_heap():
    def __init__(self):
        self.heap_value,self.heap_index=[[],[]]
        #self.heap=[self.heap_value,self.heap_index]
        self.capacity=1
        self.size=0

# Add new values to the heap without sorting them 
    def add_heap(self,new_value): 
        self.heap_value.append(new_value)
        self.heap_index.append(self.size+1) # Index starts with 1
        self.size+=1
        while self.size>self.capacity:
            self.capacity=2*self.capacity
        self.sortit()
            
# Remove values from heap, by swapping with the last index element. 
 
    def remove(self,old_value): 
        where_at=self.heap_value.index(old_value)
        temp=self.heap_value[where_at]
        self.heap_value[where_at]=self.heap_value[self.size-1]
        self.heap_value[self.size-1]=temp
        del temp
        del self.heap_value[self.size-1]
        self.heap_index.remove(self.size)
        self.size+=-1
        print("Item removed",old_value)
        self.sortit()
#         self.printall()
        
# Print the last value of heap
    def peek(self):
        print("Peek top value is :", self.heap_value[self.size-1])
        return self.heap_value[self.size-1]

# Remove the first/top value on heap
    def pop(self):
        print("Top value being poped is :", self.heap_value[0])
        val=self.heap_value[0]
        self.remove(val)
        return val
    
# Print all elements
    def printall(self):
        count=0
        binary=0
        for i in range(self.size):
            temp=self.heap_value
            print(temp[i]," ", end ="")
            count+=1
            if count>=(2**binary):
                binary+=1
                count=0
                print("\n", end ="")
            
        print("\nTotal No of items is:", self.size)
        
# Check every element to its parent, starting from the bottom of heap. If parent is larger, swap. 
# Repeat until smallest is on top of heap
    def swim(self, index): 
        
        current=index
        parent=int(index/2 -0.5)
        
#         print("Index:",index," Parent:",parent)
#         print("Index value:",self.heap_value[index]," Parent value:",self.heap_value[parent])
        change=False
        
        level=int(log(self.size+1,2))
        for limit in range(level):
            
            while self.heap_value[parent]>self.heap_value[index]:
#                 print("Swiming Up Note")
                change=True
                #(swap parent and index)
                temp=self.heap_value[parent]
                self.heap_value[parent]=self.heap_value[index]
                self.heap_value[index]=temp

                # Reassign parent index 
                index=int(index/2-0.5)
                parent=int(index/2-0.5)
                
                if log(index+1,2)<=limit:
                    break
            index=current
            parent=int(index/2 -0.5)
            
        if change:
#             self.printall()
            return

# Check every element to its children, starting from the top of heap. If parent is larger, swap. 
# Repeat until the larger element is pushed to the bottom of heap until condition met.
    def sink(self, index): 
        # This function needs work
        current=index
        child1=int(2*index+1)
        child2=int(2*index+2)
        change=False
        
        # Corner case conditions for empty heap and last of heap with only 1 child
        if child1>=self.size:
            return None
        if child1<self.size and child2>=self.size:
            child2=child1
        
        #child=int(2*index, 2*index+1)
#         print("Index:",index," child:",child1," ",child2)
#         print("Index value:",self.heap_value[index]," child value:",self.heap_value[child1],self.heap_value[child2])
        
        # Loop for checks, log2(n) times.
        level=int(log(self.size+1,2))
        for limit in range(level):
            
            # cheap if loop conditions are met, else select child to swap with
            while self.heap_value[index]>self.heap_value[child1] or self.heap_value[index]>self.heap_value[child2] :
#                 print("Sink Up Note")
                change=True
                
                if self.heap_value[index]>self.heap_value[child1] and self.heap_value[index]>self.heap_value[child2]:
                    if self.heap_value[child1]<=self.heap_value[child2]:
                        child=child1
                    else:
                        child=child2                    
                elif self.heap_value[index]>self.heap_value[child1]:
                     child=child1
                elif self.heap_value[index]>self.heap_value[child2]:
                     child=child2
                        
                # Swap parent index to repeat until bottom of heap
                temp=self.heap_value[child]
                self.heap_value[child]=self.heap_value[index]
                self.heap_value[index]=temp
            
                 # Check if bottom of heap or only 1 child
                index=child
                child1=int(2*index+1)
                child2=int(2*index+2)
                if child1 >=self.size:
                    break
                elif child2 >=self.size:
                    child2=child1              
#                 self.printall()
                
            index=current
            child1=int(2*index+1)
            child2=int(2*index+2)
            if child1>self.size and child2>self.size:
                break
            if child1<self.size and child2>=self.size:
                child2=child1
               
        if change:
#             self.printall()
            return

# Sort entire heap using swim and sink functions when elements are removed from heap. 
# ************************************* #
# ******************* Note ************ #
# ************************************* #
#: Either swim or sink function is required. Not both. 

    def sortit(self):
        for index in range(self.size-1,0,-1):
            self.swim(index)
        for index in range(0,self.size):
            self.sink(index)
        #Create a map for keeping track of elements
        #heapify cs.umd.edu
```

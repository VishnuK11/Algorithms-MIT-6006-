# Binary Sorted Tree (Unbalanced)

```python
# BST - A binary list with smaller on left or bigger on right. But unbalanced
# Condition of heap: value(parent) < value(child 1) and value(parent) < value(child 2)
import random
from math import log
```

```python
class bst:
    # Adding
    # Removing
    # Printing
    # Traversing
    
    def __init__(self):
        self.value=[]
        self.capacity=1 # starts with 1
        self.size=0  # starts with zero
#         self.last=0  # starts with zero
        self.value.append(None)
        self.sorted=[]
        
    # Add new elements into heap    
    def add(self,add_value):
        current=0
        added=True
        # Find Slot
        while self.value[current]!=None:
            if add_value<self.value[current]:
                current=2*current+1
            elif add_value>self.value[current]:
                current=2*current+2
            else:
                print("Duplicate Found. Not added")
                added=False
                return None # Duplicate handling - Create another copy on left
            if current>=self.capacity:
                self.add_row()
        self.value[current]=add_value
        self.size+=1
#         self.printall()  
      
    # Add new row when heap is full
    def add_row(self): # Add a new row of elemments and initiate it to none
        rows=log(self.capacity+1,2)
        self.capacity+= int(pow(2,rows))
        for i in range(int(pow(2,rows))):
            self.value.append(None)
            
    def pop(self):
        val=self.value[0]
        self.remove(val,-1)
        return val
    
    def remove(self,remove_value,at):
        if at==-1:
            current=0
        else:
            current=at
        case=-1
        # Find Slot   # Check Case 0, 1, 2, 3, 4   # Perform Case 0, 1, 2, 3, 4
        while self.value[current]!=None and current<self.capacity:
            if remove_value<self.value[current]:
                current=2*current+1
            elif remove_value>self.value[current]:
                current=2*current+2
            elif remove_value==self.value[current]:
                break
                   
        case=self.isleaf(current)
        if case==-1:
            print("Couldn't find item")
        elif case==5:
            print("Some error finding case")
        self.printall()  
       
    # Function to move subtree and check various condition of leaf elements
    def isleaf(self,current):
        child1=2*current+1
        child2=2*current+2
        #Check what happends when you remove the last most node. Might have an error trying to find it
        if  current>=self.capacity :
            print("Item not found")
            return 0 # Case 0 Item not found
        elif self.value[current]==None:
            print("Item not found")
            return 0 # Case 0 Item not found
        elif child1>=self.capacity and child2>=self.capacity:
            self.value[current]=None
            return 1 # Case 1 Leaf Node. No children
        elif self.value[child1]==None and self.value[child2]==None:
            self.value[current]=None
            return 1 # Case 1 Leaf Node. No children
        elif self.value[child1]!=None and self.value[child2]==None:
            self.value[current]=self.value[child1]
            self.remove(self.value[child1],child1)
            #             self.value[child1]=None
            #copy algorithm
            return 2 # Case 2 Left child only
        elif self.value[child1]==None and self.value[child2]!=None:
            self.value[current]=self.value[child2]
            self.remove(self.value[child2],child2)
#             self.value[child2]=None
            return 3 # Case 3 Right child only
        elif self.value[child1]!=None and self.value[child2]!=None:
            self.copy_subtree(current)
            return 4 # Case 4 Both Children
        else:
            return 5
            
    # Copy Subtree
    def copy_subtree(self, current):
        print("Doing subtree")
        p=2*current+2 # Choose right subtree
        q=2*p+1 #(Find the smallest of right subtree's left children)
        #Find Smallest in right subtree or largest in left subtree. Going with 1st one
        while q<self.capacity and self.value[q]!=None:
            p=q
            q=2*q+1
        self.value[current]=self.value[p]
        self.remove(self.value[p],p)
            
    # Print Tree as is
    def printall(self):
        count=0
        binary=0
        print("\nCapacity",self.capacity)
        for i in range(self.capacity):
            print(self.value[i]," ", end ="")
            count+=1
            if count>=(2**binary):
                binary+=1
                count=0
                print("\n", end ="")
        print()
     
    # Print Tree as is. Not working properly. Use print all
    def printtree(self):
        level=int(log(self.capacity+1,2))
        binary=0
        count=0
        print("\nCapacity",self.capacity)
        for i in range(self.capacity):
            j=int(pow(2,level-1-binary))
            for k in range(int(j/2+1)):
                print("  ", end ="")
            if self.value[i]==None:
                print("__", end ="")
            else:
                print(self.value[i],end ="")
            for k in range(int(j/2+1)):
                print("  ", end ="")
                    
            count+=1
            if count>=(2**binary):
                binary+=1
                count=0
                print("\n", end ="")
        
    # Use dynamic programming to print in order. Repeat: Left - Start - Right:
    def print_inorder(self,start):
        left=2*start+1
        right=2*start+2
        if start==0:
            self.sorted=[]
        if start<self.capacity and self.value[start]!=None:
            self.print_inorder(left)
            self.sorted.append(self.value[start])
            self.print_inorder(right)
        else:
            return
        if start==0:
            print(self.sorted)
```
